```python




comb =  tweets[['user', 'likes', 'timestamp']].groupby('user').agg(({'likes':'mean', 'timestamp':'count'})).reset_index().sort_values(by='timestamp', ascending=False)
tw = tweets.sort_values(by='likes', ascending=False).head(1000)
```

```python
comb[(comb['likes'] > 100.0) & (comb['timestamp'] > 50)].sort_values(by='likes', ascending=False)
```

```python
tw_count =  tweets[['user', 'timestamp']].groupby(['user']).count().sort_values(by='timestamp',ascending=False)

tw = tw.merge(tw_count, left_on='user', right_on='user')
```

```python
tw_f = tw[tw.timestamp_y > 100]
```

```python
users  = pd.Series(pd.unique(tw_f.user))
```

### Bitcoin new discovery
Filter by followers and number of tweets, filter also by whether decrption contains the word bitcoin.

```python
tweet_file = '/Users/dimmehal/Desktop/work/Twitter/Bitcoin_tweets.csv'

tweets = pd.read_csv(tweet_file, sep=',')
```

```python
agg_tweets = tweets[['user_name', 'user_description','user_followers', 'date']].groupby(['user_name', 'user_description']).agg({'date':'count', 'user_followers':'max'})

agg_tweets = agg_tweets[(agg_tweets['date'] > 50) &( agg_tweets['user_followers'] > 10000)]
agg_tweets = agg_tweets.reset_index()
filtered_agg_tweet = agg_tweets
#filtered_agg_tweet = agg_tweets[agg_tweets['user_description'].astype(str).str.contains('bitcoin|Crypto|btc') == True]
filtered_agg_tweet
```
